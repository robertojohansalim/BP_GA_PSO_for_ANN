```python
import matplotlib.pyplot as plt
import pickle
import os

def compile_report(datasets_name="IRIS"):
    PATH = "./" + datasets_name
    plt.figure(figsize=(15,10))
    for file_name in os.listdir(PATH):
        if file_name.endswith(".pkl"):
            report = pickle.load(open(PATH + "/" + file_name, "rb"))
            accuracy_history = report["accuracy_history"]
            if max(accuracy_history) <= 1:
                print("Readjust ", file_name)
                accuracy_history = [acc * 100 for acc in accuracy_history]
            plt.plot(accuracy_history, label=file_name)
    plt.legend()
```

```python
datasets_name = "IRIS"
compile_report(datasets_name)
```

```python
datasets_name = "WINE"
compile_report(datasets_name)
```

```python
datasets_name = "DIGITS"
compile_report(datasets_name)
```

```python
datasets_name = "MNIST"
compile_report(datasets_name)
```
